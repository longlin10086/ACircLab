```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']=['Source Han Serif CN']
plt.rcParams['axes.unicode_minus']=False


# 频率和放大倍数数据
frequencies = [10, 50, 100, 1000, 10000, 100000, 1000000, 2000000, 3000000]
ui = [0.7, 0.7, 0.7, 0.7, 0.7, 0.7, 0.7, 0.7, 0.7]
uo = [0.3647, 0.6701, 0.6993, 0.6827, 0.6819, 0.6712, 0.6952, 0.6835, 0.6729]
amplification_factors = [uo[i] / ui[i] for i in range(len(ui))]

plt.figure(figsize=(10, 6))
plt.plot(frequencies, amplification_factors, marker='o', linestyle='-')
plt.xscale('log')
plt.xlabel('频率 (Hz)')
plt.ylabel('放大倍数 Au')
plt.title('射极跟随器的幅频响应特性')
plt.grid(True)

plt.savefig('射极跟随器的幅频响应特性.png')
plt.show()
```

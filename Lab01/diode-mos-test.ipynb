# 实验一 二极管与三极管的功能测试

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family']=['Source Han Serif CN']
plt.rcParams['axes.unicode_minus']=False
```

## 5 实验数据分析

1. 由实验数据表 1-4，做出 $I_B=f(U_{BE})|U_{CE=0}$ 和 $I_B=f(U_{BE})|U_{CE=2V}$ 特性曲线，并分析差别的原因

```python
Ib = np.array([0, 2, 4, 6, 10, 20, 40, 60, 80]) # (uA)
Ube_1 = np.array([0, 0.501, 0.53305, 0.54412, 0.56137, 0.58034, 0.59837, 0.61064, 0.61878]) # Uce=0V (V)
Ube_2 = np.array([0, 0.6029, 0.6263, 0.6369, 0.6459, 0.6660, 0.6847, 0.6923, 0.6980]) # Uce=2V (V)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.plot(Ube_1, Ib, 'ro-')
plt.ylabel(r'$I_B$ / $\mu A$')
plt.xlabel(r'$U_{BE}$ / V')
plt.title(r'晶体管的输入特性曲线($U_{CE}=0V$)')
plt.grid(True)

plt.subplot(122)
plt.plot(Ube_2, Ib, 'bo-')
plt.ylabel(r'$I_B$ / $\mu A$')
plt.xlabel(r'$U_{BE}$ / V')
plt.title(r'晶体管的输入特性曲线($U_{CE}=2V$)')
plt.grid(True)

plt.savefig('晶体管的输入特性曲线.png')
plt.show()
```

当 $U_{CE}\ge 1V$ 时，晶体管处于放大状态，基极电流的变化受 $U_{BE}$ 控制

2. 由实验数据表 1-5，在同一个坐标系下，分别做出三条 $I_C=f(U_{CE})|I_{B=\text{常数}}$ 的特性曲线，并分析曲线的原因

```python
Uce = np.array([0, 1, 2, 3, 5, 10]) # (V)
Ic_1 = np.array([0.0683, 2.4273, 2.4380, 2.9064, 2.9183, 3.0953]) # Ib=10uA (mA)
Ic_2 = np.array([0.1209, 6.0423, 6.1427, 6.2330, 6.5553, 7.2675]) # Ib=20uA (mA)
Ic_3 = np.array([0.1785, 8.8490, 9.572, 9.8523, 10.0578, 11.7573]) # Ib=30uA (mA)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(Uce, Ic_1, 'ro-')
plt.plot(Uce, Ic_2, 'co-')
plt.plot(Uce, Ic_3, 'yo-')
plt.legend([r'I_b=10$\mu A$', r'I_b=20$\mu A$', r'I_b=30$\mu A$'], loc='upper left')
plt.xlabel(r'$U_{CE}$ / V')
plt.ylabel(r'$I_C$ / mA')
plt.title(r'晶体管的输出特性曲线')
plt.grid(True)

plt.savefig('晶体管的输出特性曲线.png')
plt.show()
```

当 $U_{CE}$ 比较小时，晶体管工作在饱和区，此时输出特性曲线处在上升区；当 $U_{CE}$ 逐渐增大至某一值后，三极管工作在放大区，此时输出特性曲线近乎水平且间距较均匀

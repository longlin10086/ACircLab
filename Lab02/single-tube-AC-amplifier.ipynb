```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family']=['Source Han Serif CN']
plt.rcParams['axes.unicode_minus']=False
```

# 表 2-3

```python
# 模拟输入信号 ui
time = np.linspace(0, 2 * np.pi, 1000)
ui = np.sin(time)

# 模拟输出信号 uo
uo_with_Ce = 2.186 * np.sin(time)  # Ce = 47μF
uo_without_Ce = 132.77 * np.sin(time)  # Ce 断开

plt.figure(figsize=(12, 6))

plt.plot(time, ui, label='输入信号 $u_i$', linestyle='--')
plt.plot(time, uo_with_Ce, label='$Ce=47μF$ 的输出信号 $u_o$')
plt.plot(time, uo_without_Ce, label='$Ce$ 断开的输出信号 $u_o$')

plt.title('Ce 对放大倍数的影响')
plt.xlabel('时间')
plt.ylabel('幅度')
plt.legend()
plt.grid(True)

plt.savefig('Ce对放大倍数的影响.png')
plt.show()
```

# 表 2-5

```python
# 模拟输入信号 ui
time = np.linspace(0, 2 * np.pi, 1000)
ui = 0.1 * np.sin(time)  # 假设输入信号幅值为 0.1V

# 模拟输出信号 uo
uo_rp_suitable = 0.6643 * np.sin(time)  # RP 合适
uo_saturation = 0.583 * np.sin(time)  # 静态工作点接近饱和区

plt.figure(figsize=(12, 6))

plt.plot(time, ui, label='输入信号 $u_i$', linestyle='--')
plt.plot(time, uo_rp_suitable, label='RP 合适的输出信号 $u_o$')
plt.plot(time, uo_saturation, label='静态工作点接近饱和区的输出信号 $u_o$')

plt.title('静态工作点对输出电压波形的影响')
plt.xlabel('时间')
plt.ylabel('幅度')
plt.legend()
plt.grid(True)

plt.savefig('静态工作点对输出电压波形的影响.png')
plt.show()
```
